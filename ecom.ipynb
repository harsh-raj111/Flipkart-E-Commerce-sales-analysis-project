```python
import pandas as pd
data = pd.read_csv("flipkard.csv")
data.info()
```

```python
data.isnull().sum()
data['size'] =  data['size'].fillna(0)
```

```python
data.isnull().sum()
```

```python
discounted_price = data['price']-(data['price']*(data['discount_percent']/100))
discounted_price
```

```python
data['Total_revenue'] = discounted_price * data['units_sold']
data['Total_revenue'].sum()
```

```python
data.groupby('category')['Total_revenue'].sum().sort_values(ascending=False)
```

key insights:
- toys generate the highest total revenue, followed closely by beauty and fashion.
- the revenue difference between categories is relatively small.
- this indicates that revenue is evenly distributed across categories , with no single category overwhelmingly dominating sales.
- the bsuiness appears to have a balanced revenue structure across product segments

```python
data.columns
```

```python
data['product_revenue'] = data['final_price'] * data['units_sold']
data.groupby('product_name')['product_revenue'].sum().sort_values(ascending=False).head(10)
```

key insights:

- The highest revenue generating product is HP Series 622 .
- Other high-performing products include Apple , Prime 418 and Phillips ultra 569
- Revenue concentration appears among premium branded products .
- Electronics and branded appliances dominate the top revenue list.

Business insight:
High revenue products are likely driven by :
- Strong brand reputation
- Higher price points
- Higher units sold

These products should be prioritized for 
- Marketing campaigns
- Inventory Optimization
-Bundle offers

```python
seller_revenue = data.groupby('seller')['product_revenue'].sum().sort_values(ascending= False).head(10)
seller_revenue
```

Seller Revenue Analysis

- UrbanRetails generate the highest total revenue.
- Followed by Quickshop and BestBuy
- Revenue distribution across top sellers is relatively close , indicating competitive performance.

Business Insight:
High - performing sellers may have :
- Better product mix
- Higher demand categories
- Better pricing strategy
- Strong ratings or faster delivery

```python
brand_revenue = data.groupby('brand')['product_revenue'].sum().sort_values(ascending=False)
brand_revenue.head(10)
```

```python
brand_by_unitSold = data.groupby('brand')['units_sold'].sum().sort_values(ascending=False)
brand_by_unitSold.head(10)
```

```python
revenue_set = set(brand_revenue.index)
unit_set = set(brand_by_unitSold.index)
common_brands = revenue_set & unit_set
only_revenue = revenue_set - unit_set
print("Brands with both revenue and units sold data:", common_brands)
print("Brands with only revenue data:", only_revenue)
```

Brand Performance Analysis

Brands appearing in both top revenue and top units sold lists represent true market leaders.  
These brands generate high income while also maintaining strong sales volume.

No brands were found that rank high in revenue but not in units sold, indicating that revenue is primarily driven by volume rather than premium pricing.

This suggests the marketplace favors mass-market products over niche luxury items.


```python
category_unit_sold = data.groupby('category')['units_sold'].sum().sort_values(ascending=False)
category_unit_sold.head(10)
```

```python
data['price_range'] = pd.cut(data['final_price'],bins=5,labels=['Low','Medium','High','Very High','Premium'])
price_demand = data.groupby('price_range')['units_sold'].sum().sort_values(ascending=False)
price_demand
```

customers prefer mid-priced products over premium ones.
sales volume decreases as price increases, indicating price-sensitive demand.
low and medium price ranges contribute the highest number of units sold, while premium products sell the least.

```python
data['reviews'] = pd.cut(data['rating'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
review_demand = data.groupby('reviews')['units_sold'].sum().sort_values(ascending=False)
review_demand
```

```python
data['rating_revenue'] = pd.cut(data['rating'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
rating_revenue = data.groupby('rating_revenue')['product_revenue'].sum().sort_values(ascending=False)
rating_revenue
```

```python
sales_by_deliveryspeed = data.groupby('delivery_days')['units_sold'].sum().sort_values(ascending=False)
sales_by_deliveryspeed
```

Impact of Delivery Time on Sales

- products with approximately 7 days delivery generate the highest total sales volume.
- faster delivery does not necessarily lead to higher demand.
- this suggests that standard delivery times are acceptable for most customers.
- sales performance may depend more on product value than delivery speed.

total sales are influnced by the number of products offered with each delivery duration,not solely by customer preference.

```python
correlation = data.corr(numeric_only=True)
correlation
```

Correlation analysis shows that price and final_price are strongly 
positively correlated, as expected, while discount percentage has a 
moderate negative relationship with final_price.

Most other variables, including rating, review count, delivery time, 
and stock availability, show very weak correlations with units sold. 
This suggests that sales are influenced by multiple factors rather 
than a single variable

```python
units_corr = data.corr(numeric_only=True)['units_sold'].sort_values(ascending=False)
units_corr
```

Correlation analysis with units_sold shows that product_revenue has a 
strong positive relationship with sales, which is expected since revenue 
depends directly on units sold.

All other variables exhibit very weak correlations, indicating that 
no single factor such as price, rating, delivery time, or discount 
alone strongly determines sales performance. This suggests that sales 
are influenced by multiple combined factors rather than a single driver.

The analysis reveals that mid-priced products, especially in Toys and 
Electronics categories, generate the highest sales volume. Price-sensitive 
demand is evident as sales decrease in premium price ranges.

Customer ratings alone do not strongly influence sales, suggesting that 
other factors such as price positioning, product type, and availability 
play a more significant role.

Faster delivery and flexible return policies show slight positive impact 
on sales, indicating the importance of customer convenience.

Businesses should prioritize inventory and marketing efforts toward 
high-performing categories and optimize pricing strategies to maximize 
revenue and demand.
